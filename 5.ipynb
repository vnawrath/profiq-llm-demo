```python
import matplotlib.pyplot as plt
import numpy as np

# Define the x and y coordinates for the elephant trunk
x = np.linspace(-1, 1, 100)
y = np.sqrt(1 - x**2)

# Plot the elephant trunk
plt.plot(x, y, color="gray")
plt.plot(x, -y, color="gray")

# Define the x and y coordinates for the elephant ear
x = np.linspace(-1, 0, 100)
y = np.sqrt(1 - x**2)

# Plot the elephant ear
plt.plot(x - 0.5, y + 0.5, color="gray")
plt.plot(x - 0.5, -y + 0.5, color="gray")

# Define the x and y coordinates for the elephant eye
x = np.linspace(-0.2, 0.2, 100)
y = np.sqrt(0.2**2 - x**2)

# Plot the elephant eye
plt.plot(x + 0.6, y + 0.6, color="gray")
plt.plot(x + 0.6, -y + 0.6, color="gray")

# Set the axis limits
plt.xlim(-1.5, 1.5)
plt.ylim(-0.5, 1.5)

# Remove the axis ticks and labels
plt.xticks([])
plt.yticks([])

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt


def draw_circle(x, y, radius, color):
    circle = plt.Circle((x, y), radius, color=color, fill=True)
    plt.gca().add_patch(circle)


def draw_elephant():
    fig, ax = plt.subplots()
    ax.set_aspect(1)

    # Body
    draw_circle(0.5, 0.5, 0.4, "gray")

    # Head
    draw_circle(0.3, 0.6, 0.15, "gray")

    # Ears
    draw_circle(0.1, 0.65, 0.05, "gray")
    draw_circle(0.1, 0.55, 0.05, "gray")
    draw_circle(0.7, 0.65, 0.05, "gray")
    draw_circle(0.7, 0.55, 0.05, "gray")

    # Eyes
    draw_circle(0.27, 0.63, 0.02, "black")
    draw_circle(0.33, 0.63, 0.02, "black")

    # Legs
    draw_circle(0.4, 0.25, 0.05, "gray")
    draw_circle(0.4, 0.15, 0.05, "gray")
    draw_circle(0.6, 0.25, 0.05, "gray")
    draw_circle(0.6, 0.15, 0.05, "gray")

    # Tail
    draw_circle(0.8, 0.25, 0.05, "gray")
    draw_circle(0.8, 0.15, 0.05, "gray")

    # Tusks
    draw_circle(0.2, 0.4, 0.1, "white")
    draw_circle(0.2, 0.3, 0.1, "white")

    # Draw the elephant
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.axis("off")
    plt.show()


draw_elephant()
```
